### Define the function

```python
def sumUserValues(x):
    ''' This functions sums up the values entered
    by the user.'''
    
    y = input('Enter the next value. Blank to exit:')
    if y == '':
        return (float(x))
    else:
        return float(x) + (float(sumUserValues(y)))
```

### Main Program

```python
# Read first value from user.
value = input('Enter values. Blank to exit:')

if value == '':
    print('Total = 0.0')
else:
    total = sumUserValues(value)
    print()
    print('Total = %.1f' %total)
    
    
```

# Alternative Solution

```python
def sumInputValues():
    ''' Returns sum of input values.
        Type = float '''
    
    # Get input value from user.
    x = input("Enter value (blank to exit): ")
    # If blank, exit
    if x == '':
        return 0.0
    else:
        return float(x) + sumInputValues()

total = sumInputValues()
print('Total is %.1f' %total)
```
